# Getting the Fashion-MNIST dataset
If you don't already have the package you can get it with `Pkg.clone("https://github.com/JuliaML/MLDatasets.jl.git")`.

First, we load the MLDatasets package and download the dataset if it hasn't been already.  

```julia
using MLDatasets
FashionMNIST.download()
```

There's several forms of the data that you can use:

```julia
train_x, train_y = FashionMNIST.traindata();
println("Normal form of x:", size(train_x))
training_tensor = FashionMNIST.convert2features(train_x);
println("Tensor form of x:", size(training_tensor))
```

We can display a random image from the training set

```julia
n = rand(1:60000)
display(FashionMNIST.convert2image(FashionMNIST.traintensor(n)))
println("This item is in category: ", FashionMNIST.trainlabels(n))
```
